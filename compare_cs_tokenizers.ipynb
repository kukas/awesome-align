```python
from itertools import islice
N = 100000
tokenized_head_cs = list(islice(open("./data/tokenized_data/full8M.cs", "r"), N))
original_head_cs = list(islice(open("./data/original_data/full8M.cs", "r"), N))
tokenized_head_uk = list(islice(open("./data/tokenized_data/full8M.uk", "r"), N))
original_head_uk = list(islice(open("./data/original_data/full8M.uk", "r"), N))
```

```python
def detokenize(line):
    line = line.replace(" ( ", " (")
    line = line.replace(" ) ", ") ")
    line = line.replace(" )\n", ")\n")
    line = line.replace(" .", ".")
    line = line.replace(" ?", "?")
    line = line.replace(" !", "!")
    line = line.replace("\" ", "\"")
    # line = line.replace(" / ", "/")
    line = line.replace(" ,", ",")
    # line = line.replace(" : ", ":")
    line = line.replace(" - ", "-")
    return line
_tokenized_head_uk = list(map(detokenize, tokenized_head_uk))

for cs, tok_cs, orig, tok, tok_edited in zip(original_head_cs, tokenized_head_cs, original_head_uk, tokenized_head_uk, _tokenized_head_uk):
    if orig != tok_edited and (not ":" in tok):
        print(f"CZ:    {cs}")
        print(f"CZtok: {tok_cs}")
        print(f"Original: {orig}")
        print(f"TokEdite: {tok_edited}")
        print(f"Tokenizd: {tok}")
```
